```python
import pandas as pd
import matplotlib.pyplot as plt

data1 = pd.read_csv('ds1_a.csv')
y = data1['y']
x0=data1['x0']
x1=data1['x1']
n = len(y)
green_x = []
green_y = []
red_x = []
red_y = []
for i in range(n):
    label = y[i]
    if label == 0:
        green_x.append(x0[i])
        green_y.append(x1[i])
    else:
        red_x.append(x0[i])
        red_y.append(x1[i])
plt.scatter(green_x,green_y,color='green')
plt.scatter(red_x,red_y,color='red')
print(data1.head(10))
```
